```python
markdown_document = """
# Vendor Proposal

## Pricing

Basic plan costs $100.

## Security

SOC2 compliant.
"""
```

```python
from langchain_text_splitters import MarkdownHeaderTextSplitter

headers_to_split_on = [("#","h1"),("##","h2")]

splitter = MarkdownHeaderTextSplitter(headers_to_split_on=headers_to_split_on)
```

```python
docs = splitter.split_text(markdown_document)
```

```python
from rich import print

for d in docs:
    print(d.metadata)
    print(d.page_content)
```

### Why Production Systems Need This

#### Headers contain:

* human semantic grouping
* contextual hierarchy
* retrieval metadata

#### You should preserve:

* section name
* subsection

* table references
